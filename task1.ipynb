Mathematical and Statistical Operations
 
load data, skipping deader

```python
raw_data = np.genfromtxt("data.txt", delimiter=",", skip_header=1, dtype=str)
```

  Extract 'Value' column and convert to integer

```python
values = raw_data[:,2].astype(int)
print(values)
print(f"Successfully loaded {len(values)} values.\n")
```

```python
print("Performing operations on 'Value' column.")
```

```python
mean_val = np.mean(values)
median_val = np.median(values)
std_val = np.std(values)
min_val = np.min(values)
max_val = np.max(values)
sum_val = np.sum(values)
print(f"Mean: {mean_val:.2f}")
print(f"Median: {median_val}")
print(f"Standard Deviation: {std_val:.2f}")
print(f"Min: {min_val}")
print(f"Max: {max_val}")
print(f"Sum: {sum_val}")
```

Save/Load Operations

```python
print("\n3. Save/Load Operations")
output_file = "processed_values.npy"
np.save(output_file, values)
print(f"Saved values to {output_file}")
```
